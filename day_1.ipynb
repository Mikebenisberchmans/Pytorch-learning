## **Tensor operations** (manipulation)
 * Addition
 * subtraction
 * multiplication (element wise)
 * division
 * matrix_multipliction

```python
## create tensor
tensor=torch.tensor([1,2,3])
```

```python
## addition
tensor+10
```

```python
## multiplication
tensor*10
```

```python
## subtraction
tensor-10
```

```python
## built in torch method
torch.mul(tensor,10)
```

```python
%%time
## matrix multiplication
torch.matmul(tensor.type(torch.long),torch.tensor([1,2,3],dtype=torch.long))
```

```python
%%time
val=0
for i in range(len(tensor)):
  val+=tensor[i]*tensor[i]
```
