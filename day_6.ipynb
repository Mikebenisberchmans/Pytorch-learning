## **Dependencies**

```python
import torch
from torch import nn
import zipfile
import requests as re
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
## setup device agnostic code

device="cuda" if torch.cuda.is_available() else "cpu"
device
```

```python
## setting up the paths

data_path=Path('data/')
image_path=data_path/"pizza_steak_sushi"
```

```python
if image_path.is_dir():
    print(f"{image_path} directory already exists")
else:
    print(f"creating {image_path} directory")
    image_path.mkdir(parents=True)
```

```python
## download and write the data

with open(data_path/"pizza_steak_sushi.zip",mode="wb") as f:
  print(f"downloading the zip file ....")
  requests=re.get("https://github.com/mrdbourke/pytorch-deep-learning/raw/main/data/pizza_steak_sushi.zip")
  f.write(requests.content)

```

```python
## unzip the data
with zipfile.ZipFile(data_path/"pizza_steak_sushi.zip","r") as f:
  print("extracting the data...")
  f.extractall(image_path)
```

```python
filescount=0
import os
for dirpath,dirnames,filenames in os.walk(data_path):
  print(f"there are {len(dirnames)} directories and {len(filenames)} files in {dirpath}.")
  filescount+=len(filenames)
```
